# 1: Defina en sus palabras los 4 paradigmas de la programación orientada a objetos y de un ejemplo funcional de cada uno. 


La programación orientada a objetos incluye cuatro pilares principales: 

**1. Abstracción:** Es el proceso de modelar, programar y escribir elementos de la vida real mediante clases y atributos. Es decir, es la acción de aislar los componentes esenciales de un objeto de aquellos que no lo son, reduciendo la complejidad del codigo, de esta manera el codigo se hace mantenible y fácil de entender. Así mismo, nos da la capacidad de reutilizar código.  

Esto se puede hacer una una analogía con el mundo real para el celular, dado que es un dispositivo complejo del cual probablemente no tengamos conocimiento de lo que ocurre en su interior cada vez que hacemos uso de una aplicación, sin embargo, sabemos como usarlo. 







```python
# Ejemplo de abstracción 

class Coche(object):
    
    # Establecemos los atributos de la clase que hemos creado.
    def __init__(self):
        # Constructor
        self.color = "Verde"
        self.doors = 4
        self.make = "Estados Unidos"
        self.model = "Sedan"
        self.estado = "Detenido"
        self.destino = " "
        self.giro = " "

    def arrancar(self):
        if self.estado == "Detenido":
            self.estado = "El carro esta en movimiento"
            print("El carro está en movimiento")
        else:
            pass 

    def ir(self):
       if self.estado == "El carro esta en movimiento":
           destino = input("Cual es el destino: ")
           print("El destino es: " + destino)
           
       else:
           print("El carro aun no esta en movimiento")
           
    def girar(self):
        if self.estado == "El carro esta en movimiento":
            giro = input("A donde desea girar?: " )
            print("El carro está girando hacia: " + giro)
            
    def parar(self):
        if self.estado == "El carro esta en movimiento":
            print("El carro se ha detenido")
        else:
            print("El carro no se ha movido")
        
if __name__ == "__main__":
    car = Coche()
    car.arrancar()
    car.ir()
    car.girar()
    car.parar()
```

**2. Herencia:** La herencia nos permite definir subclases a partir de una clase ya definida. Es decir, nos permite crear clases nuevas a partir de clases preexistentes. En la cual se heredan atributos y métodos de la clase madre. 

Pueden haber varios tipos de herencia: 
- Simple: Cuando una clase hereda de solo una clase base 

- Multiple: Es cuando una clase heresa de dos o mas clases base 

Para entender mejor este paradigma, se realizara un ejemplo en codigo. 

```python
class Persona(): #Clase madre o superclase

   def __init__(self,n,c):
                self.nombre = n
                self.id = c

   def getInfo(self):
     return (f"nombre:{self.nombre}, id:{self.id}")


class Cliente(Persona): # Clase hija o subclase, la cual hereda atributos de la clase madre
  tipo = "" 
  def __init__ (self,n,c,tipo):
    super().__init__(n,c)
    self.tipo= tipo
  
  def getInfo2(self):
    info= super().getInfo()+(f", Tipo: {self.tipo}")
    return (info)

if __name__=="__main__":
  c1=Cliente("Luna Escamilla",12345,"VIP")
  print(c1.getInfo2())
  p1=Persona("Juan Cruz", 28)
  print(p1.getInfo())

```

**3. Polimorfismo:** Nos permite modificar ligeramente los métodos y atributos de las subclases previamente definidas en la super clase. Es decir, es la capacidad de tomar más de una forma, dado que una operación puede presentar diferentes comportamientos en diferentes instancias. Por ejemplo, objetos de diferentes clases pueden ser accedidos utilizando el mismo interfaz, mostrando un comportamiento distinto según como sean accedidos. El polimorfismo es ampliamente utilizado en la aplicación de la herencia. 

```python
#Ejemplo polimorfismo

#Tenemos una clase Animal con un método hablar. 

class Animal:
    def hablar(self):
        pass
#Por otro lado tenemos otras dos clases que heredan de la anterior, Además implementan el método hablar de una forma distinta

class Perro(Animal):
    def hablar(self):
        print("Guau!")

class Gato(Animal):
    def hablar(self):
        print("Miau!")

#Creamos un objeto de cada clase y llamamos al método hablar. Aquí podemos observar el polimorfismo dado que cada animal se comporta de manera distinta al usar la clase madre. 

for animal in Perro(), Gato():
    animal.hablar()
```

# 2: Consulte qué es la modularidad en la programación y cómo ayuda esto al mantenimiento a largo plazo de grandes aplicaciones.

La modularidad va muy relacionada con las clases y es un principio de la Programación Orientada a Objetos. Así mismo, va relacionada con el Diseño Modular que significa dividir un sistema en partes pequeñas, las cuales serán los módulos que podrán funcionar de manera independiente.Por ejemplo, cuando se requieren resolver problemas mas complejos, en los que el número de instrucciones aumenta, existen técnicas que permiten resolver de forma eficiente estos problemas, facilitando al programador la construcción de la solución. Es decir, habría que dividir el programa en subprogramas que realicen tareas especificas. 

La modularidad de nuestro código nos va a permitir:



*   Evitar colapsos
*   Hacer nuestro código más entendible
*   Legibilidad
*   Resolución rápida de problemas
*   Dividir la complejidad de un problema convirtiendo problemas complejos en un conjunto de problemas más simples y por tanto más sencillos de implementar
*   Reutilizar el código de un programa en cualquier momento de la ejecución del mismo


En otras palabras, la modularidad, nos permite poder modificar las características de la clase que definen a un objeto, de forma independiente de las demás clases en la aplicación. Es decir, que nuestra aplicación puede dividirse en módulos separados, normalmente clases, y estos módulos pueden compilarse y modificarse sin afectar a los demás.

- **Atributos:** Son las propiedades de los elementos que se trabaja. Es decir, son las caracteristicas de un objeto, que ayudan a poder definir su estructura y permite diferenciarlo de otros objetos

- **Métodos:** Son las acciones que pueden hacer los objetos; son los comportamientos o conductas de un objeto y con las cuales se permite identificar la forma en que actua con respecto a su entorno y a otros objetos.

```python
# Este es un ejercicio hecho en clase
# se aplica todo lo relacioando con clases, metodos, atributos, constructor e incluso herencias 

import numpy as np

# Creo la clase padre, que va a contener el atributo y los metodos del programa.
class Shape:

  #Defino el atributo de la clase padre
    def __init__(self, color):
        self.c = color

  #Defino los metodos que se harán - Luego se van a sobreescribir, acorde a 
  #lo que se necesite.
     
    def calculatePerimeter(self):
        pass
    
    def calculateArea(self):
        pass

#Creo la clase hija para rectangulo       
class Rectangle(Shape):
    
    #Definimos los atributos que va a recibir la clase
    def __init__(self, color, width, height):
        self.c = color
        self.w = width
        self.h = height
        
    #Se definen los metodos de area y perimetro para el caso del rectangulo.
    def calculateArea(self):
        area = round (self.w * self.h, 2)
        return area
     
    def calculatePerimeter(self,):
        perimeter = round(2 * (self.w + self.h), 22)
        return perimeter

#Creo la clase hija para cuadrado      
class Square(Rectangle):
    
    #Definimos los atributos que va a recibir la clase cuadrado.
    def __init__(self, color, side):
        self.c = color
        self.s = side
    
    #Definimos los metodos de area y perimetro para el caso del cuadrado.
    def calculateArea(self):    
        area = round (self.s * self.s, 2)
        return area
        
    def calculatePerimeter(self,):
        perimeter = round(4 * self.s, 2)
        return perimeter

#Creo la clase hija para el circulo 
class Circle(Shape):
    
    #Definimos los atributos que va a recibir la clase circulo
    def __init__(self, color, radiun):
        self.c = color
        self.r = radiun
    
    #Definimos los metodos de area y perimetro para el caso del cuadrado.
    def calculatePerimeter(self):
        perimeter = round(2 * np.pi * self.r, 2)
        return perimeter
    
    def calculateArea(self):
        area = round(np.pi * self.r ** 2, 2)
        return area

print("Making a Rectangle")
shape1 = Rectangle("Red", 5, 3)

print("Making a Square")
shape2 = Square("Blue", 4.5)

print("Making a Circle")
shape3 = Circle("Green", 6)

print(f"The rectangle has a perimeter of {shape1.calculatePerimeter()}.")
print(f"The square has a perimeter of {shape2.calculatePerimeter()}.")

print(f"The circle has a perimeter of {shape3.calculatePerimeter()}.")
print(f"The rectangle has an area of {shape1.calculateArea()}.")

print(f"The square has an area of {shape2.calculateArea()}.")
print(f"The circle has an area of {shape3.calculateArea()}.")

print("Can I make an abstract `Shape` object? Let's see!")
shape4 = Shape("Black")

print("If I change the code so I can make the shape, let's try getting it's area and perimeter!")
print(f"Shape perimeter: {shape4.calculatePerimeter()}")
print(f"Shape area: {shape4.calculateArea()}")
```
